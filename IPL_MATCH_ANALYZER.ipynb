```python
# ---------------------------------------------
#  Required Libraries (install if needed):
#  pip install matplotlib seaborn pandas
# ---------------------------------------------

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1️⃣ Create the data
# Simulated count of matches won/lost by the toss-winning team
data = {
    "Result": ["Lost", "Won"],
    "Count": [520, 540]
}

df = pd.DataFrame(data)

# 2️⃣ Calculate percentages for labels
total = df["Count"].sum()
df["Percentage"] = (df["Count"] / total * 100).round(1)

# 3️⃣ Set up the visual style
sns.set_theme(style="whitegrid")
plt.figure(figsize=(7, 5))

# 4️⃣ Create bar plot with custom colors
colors = ['#a6cee3', '#fdbf6f']
barplot = sns.barplot(x="Result", y="Count", data=df, palette=colors)

# 5️⃣ Add title and labels
plt.title("Does Winning the Toss Increase Match Winning Chances?", fontsize=14, weight="bold")
plt.xlabel("Did the Toss Winning Team Win the Match?", fontsize=12)
plt.ylabel("Count", fontsize=12)

# 6️⃣ Annotate bars with percentages
for i, row in df.iterrows():
    barplot.text(i, row["Count"] + 10, f"{row['Percentage']}%",
                 ha='center', va='bottom', fontsize=12, weight="bold", color='black')

# 7️⃣ Display the plot
plt.tight_layout()
# plt.savefig("toss_vs_win.png", dpi=300)  # Optional: save as image
plt.show()


df = pd.read_csv("/content/drive/MyDrive/matches.csv")
print("Available Columns:", df.columns)
```

```python
# -------------------------------------------------------
# Required Libraries
# pip install pandas matplotlib
# -------------------------------------------------------

import pandas as pd
import matplotlib.pyplot as plt

# 1️⃣ Replace this with your actual DataFrame from a CSV or database
# Simulated example structure:
data = {
    'Venue': ['Wankhede Stadium', 'Wankhede Stadium', 'Wankhede Stadium',
              'Eden Gardens', 'Eden Gardens', 'Eden Gardens',
              'Narendra Modi Stadium', 'Narendra Modi Stadium', 'Narendra Modi Stadium'],
    'Team': ['Mumbai Indians', 'Chennai Super Kings', 'Delhi Capitals',
             'Kolkata Knight Riders', 'Sunrisers Hyderabad', 'Chennai Super Kings',
             'Gujarat Titans', 'Delhi Capitals', 'Mumbai Indians'],
    'Wins': [25, 18, 10, 22, 15, 13, 16, 14, 12]
}

df = pd.DataFrame(data)

# 2️⃣ Pivot the table: venues as rows, teams as columns
pivot_df = df.pivot_table(index="Venue", columns="Team", values="Wins", aggfunc='sum').fillna(0)

# 3️⃣ Plotting
plt.figure(figsize=(24, 10))  # Wide to fit many venues
pivot_df.plot(kind='bar', stacked=True, figsize=(24, 10), colormap='tab20')

# 4️⃣ Formatting the plot
plt.title("Stacked Team Wins at Different Venues", fontsize=16)
plt.xlabel("Venue", fontsize=12)
plt.ylabel("Wins", fontsize=12)
plt.xticks(rotation=90, ha='center')
plt.legend(title="Teams", bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
# Optional: Save figure
# plt.savefig("stacked_team_wins.png", dpi=300)
plt.show()
```
